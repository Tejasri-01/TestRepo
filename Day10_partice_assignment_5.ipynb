```python
%matplotlib inline
import pandas as pd
```

```python
cast=pd.read_csv('cast.csv',index_col=None)
cast.head()
```

```python
release_dates=pd.read_csv('release_dates.csv',index_col=None,parse_dates=['date'])
release_dates.head()
```

# 1. Make a bar plot of the months in which movies with "Christmas" In their title tend to be released in the USA.

```python
import pandas as pd
import matplotlib.pyplot as plt
christmas_movies_usa = release_dates[(release_dates['title'].str.contains('Christmas', case=False)) & (release_dates['country'] == 'USA')]
christmas_movies_usa['month'] = christmas_movies_usa['date'].dt.month
monthly_counts = christmas_movies_usa['month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Month')
plt.ylabel('Number of Christmas Movies Released')
plt.title('Number of Christmas Movies Released by Month in the USA')
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(12), months, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3. Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'release_dates'

# Filter movies with "Romance" in their title and in the USA
romance_movies_usa = release_dates[(release_dates['title'].str.contains('Romance', case=False)) & (release_dates['country'] == 'USA')]

# Extract the day of the week from the 'date' column (0=Monday, 6=Sunday)
romance_movies_usa['day_of_week'] = romance_movies_usa['date'].dt.dayofweek

# Define day of the week labels
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Count the frequency of movie releases for each day of the week
daily_counts = romance_movies_usa['day_of_week'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
daily_counts.plot(kind='bar', color='pink', edgecolor='black')

# Set plot labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Number of "Romance" Movies Released')
plt.title('Number of "Romance" Movies Released by Day of the Week in the USA')

# Set x-axis labels as day of the week names
plt.xticks(range(7), days_of_week, rotation=45)

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 5.On which date was each Judi Dench movie from the 1990s released in the USA?

```python
import pandas as pd

# Assuming you have a DataFrame named 'release_dates' and 'cast' containing the movie and cast information.

# Filter movies with Judi Dench in the cast and released in the USA
judi_dench_movies_usa = release_dates[(release_dates['country'] == 'USA') & (release_dates['title'].isin(cast[cast['name'] == 'Judi Dench']['title']))]

# Filter movies from the 1990s
judi_dench_movies_1990s = judi_dench_movies_usa[(judi_dench_movies_usa['date'].dt.year >= 1990) & (judi_dench_movies_usa['date'].dt.year <= 1999)]

# Display the release dates for each movie
print(judi_dench_movies_1990s[['title', 'date']])
```

# 6. In which months do films with Judi Dench tend to be released in the USA

```python
import pandas as pd

# Assuming you have a DataFrame named 'release_dates' and 'cast' containing the movie and cast information.

# Filter movies with Judi Dench in the cast and released in the USA
judi_dench_movies_usa = release_dates[(release_dates['country'] == 'USA') & (release_dates['title'].isin(cast[cast['name'] == 'Judi Dench']['title']))]

# Extract the month from the 'date' column
judi_dench_movies_usa['month'] = judi_dench_movies_usa['date'].dt.month

# Count the frequency of movie releases for each month
monthly_counts = judi_dench_movies_usa['month'].value_counts().sort_index()

# Determine the months with the highest and lowest release frequencies
most_common_month = monthly_counts.idxmax()
least_common_month = monthly_counts.idxmin()

# Display the results
print(f"Films with Judi Dench tend to be released most often in the month of {most_common_month}.")
print(f"Films with Judi Dench tend to be released least often in the month of {least_common_month}.")
```
