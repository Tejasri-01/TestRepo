# Write a python program to implement the various operations for Stack ADT

1) Push 

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = [None] * capacity
        self.top = -1  # Initialize the top pointer to -1

    def is_full(self):
        return self.top == self.capacity - 1

    def is_empty(self):
        return self.top == -1

    def push(self, item):
        if self.is_full():
            print("Stack is full. Cannot push the element.")
        else:
            self.top += 1
            self.stack[self.top] = item

    def pop(self):
        if self.is_empty():
            print("Stack is empty. Cannot pop an element.")
            return None
        else:
            item = self.stack[self.top]
            self.top -= 1
            return item

    def peek(self):
        if self.is_empty():
            print("Stack is empty. Cannot peek.")
            return None
        else:
            return self.stack[self.top]

# Example usage:
stack = Stack(7)
stack.push(1)
stack.push(2)
stack.push(3)
print("Top element:", stack.peek())  # Output: 3
stack.push(4)
stack.push(5)
stack.push(6) 
```

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = [None] * capacity
        self.top = -1  # Initialize the top pointer to -1

    def is_empty(self):
        return self.top == -1

    def push(self, item):
        if self.top < self.capacity - 1:
            self.top += 1
            self.stack[self.top] = item
        else:
            print("Stack is full. Cannot push the element.")

    def pop(self):
        if not self.is_empty():
            item = self.stack[self.top]
            self.top -= 1
            return item
        else:
            print("Stack is empty. Cannot pop an element.")
            return None

    def peek(self):
        if not self.is_empty():
            return self.stack[self.top]
        else:
            print("Stack is empty. Cannot peek.")
            return None

# Example usage:
stack = Stack(2)
stack.push(11)
stack.push(62)
stack.push(53)

# Pop the top element
popped_element = stack.pop()
print("Popped element:", popped_element)  # Output: Popped element: 3
```

```python
class Stack:
    def __init__(self):
        self.__stack = []

    def is_empty(self):
        return len(self.__stack) == 0

    def size(self):
        return len(self.__stack)

    def push(self, item):
        self.__stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.__stack.pop()
        else:
            return None  # Return None if the stack is empty

    def display(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            print("Stack Contents:")
            for item in reversed(self.__stack):
                print(item)

# Create a stack
stack = Stack()

# Push some items onto the stack
stack.push(1)
stack.push(2)
stack.push(3)

# Display the stack contents
stack.display()
```

# 6.White a Python script to perform the following operations on a singly linked list

& Create a list

b. Find the smallest element from the list

e. Insert an element if it is not a duplicate element

4 Display the elements in reverse order

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def create_list(self, elements):
        for element in elements:
            self.insert(element)

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None
        smallest = self.head.data
        current = self.head.next
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest

    def insert_unique(self, data):
        current = self.head
        while current:
            if current.data == data:
                return  # Element already exists, don't insert
            current = current.next
        self.insert(data)

    def display_reverse(self):
        def print_reverse(node):
            if node:
                print_reverse(node.next)
                print(node.data)

        print_reverse(self.head)

# Example usage:
if __name__ == "__main__":
    elements = [5, 2, 8, 2, 1, 5, 7]
    linked_list = LinkedList()
    linked_list.create_list(elements)

    print("Original List:")
    current = linked_list.head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

    smallest_element = linked_list.find_smallest()
    print("Smallest Element:", smallest_element)

    linked_list.insert_unique(3)
    linked_list.insert_unique(5)  # Duplicate, won't be inserted

    print("Updated List:")
    current = linked_list.head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

    print("List in Reverse Order:")
    linked_list.display_reverse()

```

1. If greater, move to right of list or else move to another side of the list. g. Print the result along with the position of the element.

```python
def binary_search(arr, target):
    arr.sort()  # Sort the list in ascending order
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2  # Calculate the middle index

        if arr[mid] == target:
            return f"Found {target} at position {mid}"
        elif arr[mid] < target:
            left = mid + 1  # Move to the right half
        else:
            right = mid - 1  # Move to the left half

    return f"{target} not found in the list"

# Get the number of inputs from the user
num_inputs = int(input("Enter the number of elements in the list: "))

# Initialize an empty list to store the inputs
input_list = []

# Get the inputs from the user one by one
for i in range(num_inputs):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Get the target value from the user
target_value = int(input("Enter the target value to search for: "))

# Call the binary_search function to find the target value
result = binary_search(input_list, target_value)

# Print the result along with the position of the element
print(result)
```

# Write a Python program that performs merge sort on a list of numbers:

a Divide: If the given array has zero or one element, return. Otherwise
b.Conquer: Recursively sort the two lists (left half and right half).
b. Conquer: Recursively sort the two lists (left half and right half).
a. Call the merge sort on left half. b. Call the merge sort on right half.

C. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort the left and right halves
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Merge the sorted halves back together
    sorted_arr = merge(left_half, right_half)
    return sorted_arr

def merge(left, right):
    result = []
    left_idx, right_idx = 0, 0

    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] < right[right_idx]:
            result.append(left[left_idx])
            left_idx += 1
        else:
            result.append(right[right_idx])
            right_idx += 1

    # Add any remaining elements from the left and right sublists
    result.extend(left[left_idx:])
    result.extend(right[right_idx:])
    
    return result

# Test the merge_sort function
input_list = [12, 4, 78, 35, 19, 7, 56]
sorted_list = merge_sort(input_list)
print("Original List:", input_list)
print("Sorted List:", sorted_list)
```

# Write a Python program for sorting a list of elements using selection sort algorithm:

a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

b. In the first iteration, find the smallest element in the unsorted list and place it in the sortedlist.
c.In the second iteration, find the smallest element in the unsorted list and place it in the 

correct position by comparing with the element in the sorted list. 
d. In the third iteration, again find the smallest element in the unsorted list and place it in the

correct position by comparing with the elements in the sorted list.
e. This process continues till the unsorted list becomes empty.
f Display the sorted elements

```python
def selection_sort(arr):
    sorted_list = []  # Initialize an empty sorted list

    while arr:  # Continue until the unsorted list becomes empty
        min_index = 0

        # Find the index of the minimum element in the unsorted list
        for i in range(1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i

        # Remove the minimum element from the unsorted list and append it to the sorted list
        sorted_list.append(arr.pop(min_index))

    return sorted_list

# Test the selection_sort function
input_list = [12, 4, 78, 35, 19, 7, 56]
sorted_list = selection_sort(input_list)
print("Original List:", input_list)
print("Sorted List:", sorted_list)
```

# Write a program in python to convert the following infix expression to its postfix form using push andpop operations of a Stack

a.A/B^C+D*E-F*G

b. (B^2-4*A*C)^(1/2) (100)

```python
def infix_to_postfix(expression):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}

    def has_higher_precedence(op1, op2):
        return precedence[op1] > precedence[op2]

    def is_operator(char):
        return char in "+-*/^"

    stack = []
    postfix = []

    for char in expression:
        if char.isalnum():
            postfix.append(char)  # If it's an operand, add it to the postfix expression
        elif char == '(':
            stack.append(char)  # If it's an opening parenthesis, push it to the stack
        elif char == ')':
            while stack and stack[-1] != '(':
                postfix.append(stack.pop())  # Pop operators from the stack to postfix until '('
            stack.pop()  # Pop '(' from the stack
        elif is_operator(char):
            while stack and stack[-1] != '(' and has_higher_precedence(stack[-1], char):
                postfix.append(stack.pop())  # Pop operators with higher precedence to postfix
            stack.append(char)  # Push the current operator to the stack

    while stack:
        postfix.append(stack.pop())  # Pop remaining operators to postfix

    return ''.join(postfix)

# Test the infix_to_postfix function
infix_expression_a = "A/B^C+D*E-F*G"
infix_expression_b = "(B^2-4*A*C)^(1/2) (100)"

postfix_a = infix_to_postfix(infix_expression_a)
postfix_b = infix_to_postfix(infix_expression_b)

print("Infix Expression A:", infix_expression_a)
print("Postfix Expression A:", postfix_a)

print("Infix Expression B:", infix_expression_b)
print("Postfix Expression B:", postfix_b)

```
