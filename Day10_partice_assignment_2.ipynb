# which three years of the 1930s saw the most films released?

```python
import pandas as pd

# Load the dataset (assuming it's in a CSV file named 'film_data.csv')
df = pd.read_csv('titles.csv')

# Filter the dataset to include only films from the 1930s
films_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]

# Group the data by year and count the number of films released each year
film_counts_by_year = films_1930s['year'].value_counts().reset_index()
film_counts_by_year.columns = ['Year', 'Number of Films']

# Sort the data by the number of films released in descending order
film_counts_by_year = film_counts_by_year.sort_values(by='Number of Films', ascending=False)

# Get the top three years with the most film releases
top_three_years = film_counts_by_year.head(3)

# Print the result
print(top_three_years)
```

# Plot the number of films that have been released each decade over the history of cinema.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the 'titles.csv' file into a Pandas DataFrame
titles = pd.read_csv('titles.csv')

# Assuming you have a 'year' column in your DataFrame that represents the release year of films.
# If not, you may need to preprocess your data to extract the release year.

# Extract the release year from the 'year' column and create a new 'decade' column
titles['decade'] = (titles['year'] // 10) * 10

# Group the DataFrame by 'decade' and count the number of films in each decade
film_counts = titles['decade'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(film_counts.index, film_counts.values, width=5)
plt.xlabel('Decade')
plt.ylabel('Number of Films')
plt.title('Number of Films Released Each Decade')
plt.xticks(film_counts.index, rotation=45)
plt.show()
```

# How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
import pandas as pd
cast = pd.read_csv('cast.csv')
roles_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959)]
leading_roles_actors = len(roles_1950s[(roles_1950s['n'] == 1) & (roles_1950s['type'] == 'actor')])
leading_roles_actresses = len(roles_1950s[(roles_1950s['n'] == 1) & (roles_1950s['type'] == 'actress')])
print(f"Number of leading roles (n=1) for actors in the 1950s: {leading_roles_actors}")
print(f"Number of leading roles (n=1) for actresses in the 1950s: {leading_roles_actresses}")
```
